optimizing point cloud, remove out boundary points

```python
import cv2
import numpy as np
import open3d as o3d
from open3d import geometry
from tqdm import tqdm
import os

# infs = np.load("/suedata1/Free/xkangyu/data/dynamic_endo_scene/pulling_soft_tissues/poses_bounds.npy")
# print(infs[0])
# exit
FX = 569 # endo-pulling 569.468, 580
CX = 320.0
CY = 256.0

tanX = CX / FX
tanY = CY / FX


path = "/suedata1/Free/xkangyu/data/dynamic_endo_scene/pulling_soft_tissues/predict_point_cloud_v4_offset_pair/0_30_colored.ply"
# path = "/suedata1/Free/xkangyu/data/dynamic_endo_scene/pulling_soft_tissues/predict_point_cloud_v4_offset_coord1/frame-000000_colored.ply"
# path = "/suedata1/Free/xkangyu/data/dynamic_endo_scene/stereomis_p2_5_1/predict_point_cloud_v4_offset_coord1/005300_colored.ply"
# path = "/homes/xkangyu/mycodes/splatt3r/results/full_datasets_15cases_jscas/pulling_soft_tissues/predict_point_cloud_pt_offset_coord1/frame-000000_colored.ply"
# path = "/homes/xkangyu/mycodes/splatt3r/results/full_datasets_15cases_jscas/stereomis_p2_5_1/predict_point_cloud_pt_offset_coord1/005300_colored.ply"
points = o3d.io.read_point_cloud(path)

np_pts = np.array(points.points)
np_colors = np.array(points.colors)

indx = []
for i in tqdm(range(len(np_pts))):
    coord = np_pts[i]
    if coord[2] != 0:
        if abs(coord[0]/coord[2]) > tanX or abs(coord[1]/coord[2]) > tanY:
            indx.append(i)
print(np_pts.shape)
new_pts = np.delete(np_pts, indx, axis=0)
new_colors = np.delete(np_colors, indx, axis=0)

print(new_pts.shape)

points.points = o3d.utility.Vector3dVector(new_pts)
points.colors = o3d.utility.Vector3dVector(new_colors)
# pts = geometry.PointCloud(new_pts)

o3d.io.write_point_cloud(path.replace(".ply", "_post.ply"), points)

```
